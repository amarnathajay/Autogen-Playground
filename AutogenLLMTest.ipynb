```python
from autogen import AssistantAgent
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
# Setup API key. Add your own API key to config file or environment variable
#Llama 3 via ollama
config_list = [
    {
        "model": "llama3.1:8b",
        "base_url": "http://localhost:11434/v1",
        'api_key': 'ollama',
    },
]
```

```python
# Gemini
config_list = [
    {
        "model": "gemini-pro",
        "api_key": os.environ.get("GEMINI_API_KEY"),
        "api_type": "google"
    }
]
```

```python
small = AssistantAgent(name="small model",
                       max_consecutive_auto_reply=5,
                       system_message="You should act as a student!",
                       llm_config={
                           "config_list": config_list,
                           "temperature": 1,
                       })

big = AssistantAgent(name="big model",
                     max_consecutive_auto_reply=5,
                     system_message="Act as a teacher.",
                     llm_config={
                         "config_list": config_list,
                         "temperature": 1,
                     })
```

```python
big.initiate_chat(small, message="What are you studying and what is your ambition?")
```
